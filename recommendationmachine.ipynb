```python
import pandas as pd
df = pd.read_json("tokened_wine_review.json")
```

```python
dataset = df.get(['tokened_description', 'variety', 'country', 'title', 'points', 'price', 'description'])
dataset = dataset.assign(keyword_set = dataset.tokened_description.apply(set))
dataset.head()
```

```python
# format keywords into machine readable
keywords = set(map(lambda x: x.lower(), ["Peppery", "Spicy", "Bold", "Intense","Warm","Rich","Full-bodied","Exotic","Aromatic"]))
# count the keywords which are matched.
dataset['count_keyword_in'] = dataset.keyword_set.apply(lambda x: len(keywords.intersection(x)))
# obtain the matching score and filter the score > 3, which means there're over 3 keywords are matching the requirement
dataset[dataset.count_keyword_in >= 3].sort_values(by=['count_keyword_in', 'points'], ascending=False).get(['title',  'points', 'variety', 'country', 'price', 'description'])
```

```python
keywords = """
Rich
Sweet
Luscious
Dessert-like
Honeyed
Fruity
Aromatic
Velvety
Decadent
"""
keywords = set(filter(lambda x: bool(x), map(lambda x : x.lower(), keywords.replace("\n", " ").split(" "))))
dataset['count_keyword_in'] = dataset.keyword_set.apply(lambda x: len(keywords.intersection(x)))
dataset[dataset.count_keyword_in >= dataset.count_keyword_in.max()-1].sort_values(by=['points'], ascending=False).query('country == "France"&price<40')
```
